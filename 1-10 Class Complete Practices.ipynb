```python
# Greator then Equal to (>=)
# Less then Equal to (<=)

200>=200 # yani 200 agar 200 ka brabar ya parah ho to ans True
200>=201 # Same Situation Ans True
200>=199 # Same Situation Ans True due to 2nd Number 200 se kam hai

200<=200 # Yani 200 agar 200 ka brabar ya kam ho to ans True
200<=201 # Same Situation 2nd Value hamari ziada hai jiski waja se Ans False hoga kyu kay humne less then equal to(<=) ka sign hai hai
200<=199 #same situation last wale main ans True tha or iskay under False kyu kay humne Less then Equal to(<=) ka sign hai
```

```python
# logicial Operator
    # And
    # OR
    # NOT

# First hum (OR) Smjhange
# OR ka kam yea hota hai kay puri situation main agar ek value bhi true hogi to Ans True dega warna False like:

# True or True = True
# True or False = True
# Talse or True = True
# Talse or False = False
```

```python
a = 500
b = 700
c = 1000

(a>b) or (b>c) or (c>a) # is situation main ek value hmari True hai yani ans True hoga
# F        F        T
```

```python
# Second hum And ko dakhange
# and ki situation bilkul OR ka apposite hai hai yani situation main agar ek bhi False huwa to puri situation False hojayegi

# True and True = True
# True and False = False
# False and True = False
# False and False = False

(a>b) and (b>c) and (c>a) # is situation main ek value hmari True hai yani ans False hoga
# F        F        T
```

### Continue And Break

```python
for a in range(50):
    if a == 25: # ab iskay under hum yea bol rhe hain a loop chalega lekin jab if ki condition true hogi to loop stop ho jayega
        break # yea rok deta hai
    else:
        print(a, end=" ") # yani jab tk if ki condition true nahi hogi us waqt tk else hei chalega
```

```python
for a in range(10):
    if a == 7:
        continue # Same senario yea sari condition ko lega bus 7 ko chor kay yea true wali condition ko rok kay baki sub print krega
    else:
        print(a, end=" ") # yani is answer main 7 nahi hoga
```

```python
for a in range(3):
    name = input("Enter Your name").lower()
    if name == "admin": # ab agar koi input main Admin dega to wo print nahi krega
        continue
    else:
        print(name)
```

```python
name_list = []

for a in range(5):
    name = input("Enter Your Name:").capitalize()
    if name == "Admin":
        continue
    else:
        name_list.append(name) # iskay under humne sirf empty list name hamare input ko append krwaya hai
print(name_list)
```

```python
import random

guess_number = random.randint(1,10)
attempts = 3

for a in range(3):
    guess = int(input("Guess A Number"))
    if guess_number == guess:
        print("Great guess")
        break
    elif guess > guess_number:
        print("Number is greator")
        attempts -= 1
    elif guess < guess_number:
        print("Number is less")
        attempts -= 1
print(guess_number)
```
