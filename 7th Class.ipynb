```python
a = 10 # yea humne ek variable bna lia hai int 10 se
if a >20: # yaha humne statement di hai "agar a(10) bara hai 20 se to next line wala function run kre"
    print(f"{a} is greator then 20") # Yea simple print krdega 
else: # agar if wali statement false huwi to phr Else check krega
    print(f"{a} is smaller then 20") # or yea wala print run krega
```

```python
# ab ek situation main ate hain

# Agar Tomatos 
# 10 Rs : 50 Kgs lelo
# 20 Rs : 40 Kgs lelo
# 30 Rs : 30 Kgs lelo
# 40 Rs : 20 Kgs lelo
# 50 Rs : 10 Kgs Lelo
# 50+ Rs : 0 Kgs lelo



tomatos_price = float(input("Enter a Tomatos Price"))

if tomatos_price <= 10:
    print(f"50 Kgs lelo {tomatos_price} se, Total Amount {50*tomatos_price}")
elif tomatos_price <=20:
    print(f"40 Kgs lelo {tomatos_price} se, Total Amount {40*tomatos_price}")
elif tomatos_price <=30:
    print(f"30 Kgs lelo {tomatos_price} se, Total Amount {30*tomatos_price}")
elif tomatos_price <=40:
    print(f"20 Kgs lelo {tomatos_price} se, Total Amount {20*tomatos_price}")
elif tomatos_price <=50:
    print(f"10 Kgs lelo {tomatos_price} se, Total Amount {10*tomatos_price}")
else:
    print(f"0 Kgs Price bhot ziada hai")
```

```python
# Last main humne Int/Float main input lia tha ab hum string main krange

signal_light = input("Enter A signal Light Color").lower()
if signal_light == "red":
    print(f"Stop the light is {signal_light.upper()}")
elif signal_light == "yellow":
    print(f"Get Ready the light color is {signal_light.upper()}")
elif signal_light == "green":
    print(f"Now you Can Go!, the light color is {signal_light.upper()}")
else:
    print(f"There is not light color {signal_light.upper()} in Signal lights")
```

```python
# ab hum ek step aage chalte hain

html = float(input("Enter Your HTML Marks"))
css = float(input("Enter Your CSS Marks"))
Java = float(input("Enter Your JAVA Marks"))
python = float(input("Enter Your PYTHON Marks"))
react = float(input("Enter Your REACT Marks"))

toal_marks = html+css+Java+python+react
precentage = toal_marks/500*100

if precentage >= 90:
    grade = "A+"
elif precentage >= 80:
    grade = "A"
elif precentage >= 70:
    grade = "B"
elif precentage >= 60:
    grade = "C"
else:
    grade = "FAIL"


print(f'''
====================================
          Student Mark Sheet
====================================
    Total Marks : {toal_marks}
    Percentage  : {precentage}
    Grade       : {grade}


''')
```
