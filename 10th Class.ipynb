## Coutinue and Break

```python
for a in range(5):
    if a == 5:
        break # break yani if ki condition true hoti hai to udher hei stop hojayega
    else:
        print(a)
```

```python
for a in range(10):
    if a == 4:
        continue # yea if wali condition true hogi to usko print nahi krega baki sub print honge
    else:
        print(a)
```

```python
for a in range(3):
    username = input("Enter Your name")
    if username == "admin":
        continue
    else:
        print(username)
```

```python
name = []
for a in range(20):
    user_name = input("Enter Your Name")
    if user_name == "admin":
        continue
    else:
        name.append(user_name)
print(name)
```

```python
num = int(input("enter a num"))

while num <=3:
    print(num)
```

```python
pas = "python123"
attemps = 3
while attemps > 0:
    user_input = input("enter password")
    if user_input == pas:
        print("access granted")
        break
    else:
        attemps -=1
        print("wrong password")

        
if attemps == 0:
    print("account blocked")
```
