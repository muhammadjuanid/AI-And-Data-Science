```python
# operators

    # Arithmethic operators
        # Addation (+)
        # subraction (-)
        # multliplication (*)
        # exponands (**)
        # division two types
                # float division (/)
                # florr division (//)
        # modulues (%)

    # Relational/Compression operators
        # double Equal to (==)
        # Not Equal to(!=)
        # Greator Then(>)
        # Less Then(<)
        # Greator then equal to (>=)
        # less then equalto(<=)

    # logicial operators
        # Logical and
        # Logicial Or
        # logicial Not

    # Operator precendence is ka matlab ka kis sign main periorty kisko deni hai solve krne main
        # pemdas

    # Incremenr or Decerement operators
```

```python
# Arthmethic operators

print(200+100) # yea normal plus krne jesa hai
print(200-100) # yea Minus jesa hai
print(50*5) # yea multiply hai
print(5**3) # yea exponands hai yani base example (5 X 5 X 5)
```

```python
# Divide yea do types kay hote hain
    # Float Division (/)
    # Florr Division (//)

print(200/7) #isaky under divide(/) ka ek sign hai
print(200//7) #iskay under divide(//) kay 2 sign hai
```

```python
# Modelus(%)
    # Yea values ko divide kr kay remainder return krega
print(15%7) # Iskay under yea 15 ko divide krga 7 yani 7X2=14, Remainder 1
print(14%7) # iskay under yea total clear hojayega remainder 0
```

```python
# Relational/Compression Operators
    # Equal to(==) and Not Equal to(!=)

500==200 # Equal to lekin is sign (==) kyu kay single(=) hum assign krne kay lie krte hain ab yea ans boolean(bool) main return krega
500==500 # same as previous but Ans is True
500!=200 # uper wali situation lekin Ans True kyu kay humne not Equal ka sign lga dia hai
500!=500 # # ab yea equal hai lekin humne sign not Equal(!=) ka lgya hai
```

```python
# Greator then(>)
# Less then(<)

200>150 #kia 200 150 se barah hai Ans True ab isko smjhne ka asan tariqa jis taraf sign open hai wo value bari hogi
200>201 # iskay under Ans False hai kyu kay sign 200 ki taraf open hai magar 2nd value 201 barah hai

200<150 # Same situation magar sign less then(<) ka hai Ans False
200<201 # Same Situation Ans False
```

```python
# Greator then Equal to (>=)
# Less then Equal to (<=)

200>=200 # yani 200 agar 200 ka brabar ya parah ho to ans True
200>=201 # Same Situation Ans True
200>=199 # Same Situation Ans True due to 2nd Number 200 se kam hai

200<=200 # Yani 200 agar 200 ka brabar ya kam ho to ans True
200<=201 # Same Situation 2nd Value hamari ziada hai jiski waja se Ans False hoga kyu kay humne less then equal to(<=) ka sign hai hai
200<=199 #same situation last wale main ans True tha or iskay under False kyu kay humne Less then Equal to(<=) ka sign hai
```

```python
# logicial Operator
    # And
    # OR
    # NOT

# First hum (OR) Smjhange
# OR ka kam yea hota hai kay puri situation main agar ek value bhi true hogi to Ans True dega warna False like:

# True or True = True
# True or False = True
# Talse or True = True
# Talse or False = False
```

```python
a = 500
b = 700
c = 1000

(a>b) or (b>c) or (c>a) # is situation main ek value hmari True hai yani ans True hoga
# F        F        T
```

```python
# Second hum And ko dakhange
# and ki situation bilkul OR ka apposite hai hai yani situation main agar ek bhi False huwa to puri situation False hojayegi

# True and True = True
# True and False = False
# False and True = False
# False and False = False

(a>b) and (b>c) and (c>a) # is situation main ek value hmari True hai yani ans False hoga
# F        F        T
```
