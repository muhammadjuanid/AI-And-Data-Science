#### Q10: Count Vowels in String

```python
text = "Data science is an interdisciplinary field that uses statistics, programming, and domain knowledge to extract meaningful insights from data. It involves collecting, cleaning, and analyzing structured and unstructured datasets to identify patterns, trends, and relationships. Data scientists use tools such as Python, R, and machine learning algorithms to build predictive models and support decision making. The field is widely applied in industries like healthcare, finance, marketing, and technology, helping organizations improve efficiency and innovation. As data continues to grow rapidly, data science plays a crucial role in turning raw information into valuable knowledge for solving real world problems today worldwide.".lower()
vowels = "aeiou"
count = 0
for a in text:
    if a in vowels:
        count +=1
print(count)
```

#### Q11: Reverse List using Loop

```python
listd = [1,2,3,4,5]
reverse_list = []
for a in range(len(listd) -1, -1, -1):
    reverse_list.append(listd[a])

print(reverse_list)
```
